#### Задание 1. Соберите всю информацию о студентах в одну таблицу df. В получившейся таблице должна быть информация и оценки всех студентов из всех групп. Напечатайте несколько строк таблицы для демонстрации результата.¶

```python
import zipfile
from zipfile import ZipFile
zipFile = zipfile.ZipFile('Data.zip', 'r')
files = zipFile.namelist()
print(*files, sep='\n')
```

```python
df = None
with ZipFile("Data.zip") as myzip:
    for i in files[1:]:
        if 'marks' in i:
            if df is None:
                data = myzip.open(i)
                df = pd.read_csv(data)
            else:
                data = myzip.open(i)
                df = df.append(pd.read_csv(data))
df.set_index(df["index"], inplace=True)

new_df = None
with ZipFile("Data.zip") as myzip:
    for i in files[1:]:
        if 'info' in i:
            if new_df is None:
                data = myzip.open(i)
                new_df = pd.read_csv(data)
            else:
                data = myzip.open(i)
                new_df = new_df.append(pd.read_csv(data))

new_df.set_index(new_df["index"], inplace=True)
new_df = new_df.drop("index", axis=1)

df = df.join(new_df)
df
```

**Задание 12. Выполните задание 11 с помощью циклов. Сравните время выполнения.**

```python
%%time
from collections import defaultdict

races = defaultdict(list)
for ind, row in df.iterrows():
    cur_race = row["race/ethnicity"]
    races[cur_race].append(row["reading score"])
for i in sorted(races.keys()):
    print(i, sum(races[i]) / len(races[i]))

print()

levels = {}
for ind, row in df.iterrows():
    cur_ed = row["education"]
    if cur_ed not in levels:
        levels[cur_ed] = 101
    else:
        levels[cur_ed] = min(levels[cur_ed], row["writing score"])

for i in sorted(levels.keys()):
    print(i, levels[i])
```

#### Задание 14. Сколько студентов успешно сдали экзамен по математике?

Создайте новый столбец в таблице df под названием Math_PassStatus и запишите в него F, если студент не сдал экзамен по математике (балл за экзамен < passmark), и P иначе.

Посчитайте количество студентов, сдавших и не сдавших экзамен по математике.

Сделайте аналогичные шаги для экзаменов по чтению и письму.

```python
def decider(num):
    if num < passmark:
        return 'F'
    return 'P'

print('экзамен по математике')
df["Math_PassStatus"] = df["math score"].apply(lambda x: decider(x))
print(df["Math_PassStatus"].value_counts())
print()
print('экзамен по чтению')
df["Reading_PassStatus"] = df["reading score"].apply(lambda x: decider(x))
print(df["Reading_PassStatus"].value_counts())
print()
print('экзамен по письму')
df["Writing_PassStatus"] = df["writing score"].apply(lambda x: decider(x))
print(df["Writing_PassStatus"].value_counts())
```

#### Задание 15. Сколько студентов успешно сдали все экзамены?

Создайте столбец OverAll_PassStatus и запишите в него для каждого студента 'F', если студент не сдал хотя бы один из трех экзаменов, а иначе 'P'.

Посчитайте количество студентов, которые сдали все экзамены.

```python
def rule(row):
    if row["Math_PassStatus"] == 'P' and row["Reading_PassStatus"] == 'P' and row["Writing_PassStatus"] == 'P':
        return "P"
    return "F"

df["OverAll_PassStatus"] = df.apply(lambda x: rule(x), axis = 1)
print(df["OverAll_PassStatus"].value_counts()['P'])
```

#### Задание 16. Переведем баллы в оценки

### Система перевода баллов в оценки
####    больше 90 = A
####      80-90 = B
####      70-80 = C
####      60-70 = D
####      50-60 = E
####    меньше 50 = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

**В случае, если средний балл попадает на границу между оценками (т.е. равен ровно 60, 70 или 80 баллов), вы можете интерпретировать условие на своё усмотрение (т.е. можете поставить за 60 баллов оценку D, а можете - E).**

```python
def GetGrade(average_mark):
    if average_mark > 90:
        return 'A'
    elif average_mark >= 80:
        return 'B'
    elif average_mark >= 70:
        return 'C'
    elif average_mark >= 60:
        return 'D'
    else:
        return 'E'


df["Grade"] = df.apply(lambda x: GetGrade(np.mean((x['math score'],
                                                   x['reading score'],
                                                   x['writing score']))), 
                         axis = 1)
```
